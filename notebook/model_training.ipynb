```python
X=df.drop("math_score",axis=1)
y=df["math_score"]
```

```python
X["parental_level_of_education"]=X["parental_level_of_education"].replace({"some high school":"high school","some college":"bachelor's degree"})
```

```python
X["lunch"]=X["lunch"].replace({"free/reduced":"free"})
```

```python
for feature in X.columns:
    print(f"{feature}:{X[feature].unique()}")
```

```python
cat_col=[]
for feature in X.columns:
    if X[feature].dtype == "O":
        cat_col.append(feature)
print(cat_col)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python

y_test
```

```python
cat_col=[]
num_col=[]
for feature in X.columns:
    if df[feature].dtype =="O":
       cat_col.append(feature)
    else:
       num_col.append(feature)
```

```python
##transformation of X
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

preprocess=ColumnTransformer(transformers=[("OneHotEncoder",OneHotEncoder(drop="first"),cat_col),("StandardScaler",StandardScaler(),num_col)],remainder="passthrough")

X_train=preprocess.fit_transform(X_train)
X_test=preprocess.transform(X_test)

```

```python
print(X_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor()
reg.fit(X_train,y_train)
```

```python
reg1=LinearRegression()
reg1.fit(X_train,y_train)
```

```python
reg=RandomForestRegressor()
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
score=r2_score(y_test,y_pred)
print(score)
```

```python
y_pred=reg.predict(X_test)
y_pred = y_pred.round().astype(int)

from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score )
```

```python
y_pred=reg1.predict(X_test)
y_pred= y_pred.round().astype(int)
score=r2_score(y_test,y_pred)
print(score)
print(y_pred)
```
